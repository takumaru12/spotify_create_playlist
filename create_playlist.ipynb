```python
#アーティストのアルバムを上限50件で取得
result = sp.artist_albums('0pWR7TsFhvSCnbmHDjWgrE',limit=50)
albums_id = []
for item in result['items']:
    print(item['name'],item['id'])
    albums_id.append(item['id'])
```

```python
#曲名、ID、人気どのリストを作成、人気順にソート
songlist = []
for i in range(len(albums_id)):
    result = sp.album_tracks(albums_id[i])
    for item in result['items']:
        print(item['name'])
        print(item['id'])
        res = sp.track(item['id'])
        print(res['popularity'])
        songlist.append([item['name'],item['id'],res['popularity']])
sorted_list = sorted(songlist, key=lambda x: x[2],reverse=True)
```

```python
print(len(sorted_list))
```

```python
#指定された数を人気なものからプレイリストに追加
cnt = 90

playlist_id  = '5NXWwSHFE6GT5qVZ2437lF'

add_list = []
for i in range(cnt):
    add_list.append(sorted_list[i][1])
results = sp.user_playlist_add_tracks(username, playlist_id, add_list)
print(results)
```
